#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

norm_prices = sb_madrid.groupby('level5')['price'].apply(
    lambda x: (x - x.mean()) / x.std())

regions = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
colors = ['purple', 'red', 'orange', 'blue']

for x, y in zip(regions, colors):
    plt.hist(norm_prices[x], bins=20, alpha=0.5, label=x, color=y)

plt.xlabel('Scaled Prices') 
plt.ylabel('Number of Homes') 
plt.title('Scaled Prices in the South Belt of Madrid')
plt.show()
```

Getafe has the highest number of homes given the scaled price.

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)
 
Hint: Make a scatter plot of each population using subplots.

```python
import matplotlib.pyplot as plt

pps_data = ds[ds['level5'].isin(["Fuenlabrada", "Leganés", "Alcorcón", "Getafe"])]

ds['pps'] = pps_data['price'] / pps_data['surface']

regions =["Fuenlabrada", "Leganés",  "Getafe", "Alcorcón"]
colors = ['purple', 'red', 'orange', 'blue']

for a, b in zip(regions, colors):
    pps = ds[ds['level5'] == a]['pps']
    plt.hist(pps, bins=20, alpha=0.5, label=a, color=b)

plt.xlabel('Price Per Square Meters') 
plt.ylabel('Homes') 
plt.title('Price Per Square Meter in the South Belt of Madrid') 
plt.show()
```

```python
from ipyleaflet import Map, Marker

# ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

sb_madrid['latitude'] = sb_madrid['latitude'].str.replace(',', '.').astype(float) 
sb_madrid['longitude'] = sb_madrid['longitude'].str.replace(',', '.').astype(float)
new_df = sb_madrid[['latitude', 'longitude', 'level5']]

regions = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
list1 = []
for m in regions:
    lat_long = new_df[new_df['level5'].isin([m])]
    list1.append(lat_long)

map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20)

for city in list1:
    for (index, row) in city.iterrows():
        marker = Marker(location=[row.loc['latitude'], row.loc['longitude']], 
                        title=row.loc['level5'])
        map.add_layer(marker)


map
```
